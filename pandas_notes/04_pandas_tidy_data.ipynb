```python
ebola = pd.read_csv('./scipy-2017-tutorial-pandas-master/data/ebola_country_timeseries.csv')
```

```python
ebola.head()
```

```python
ebola_melt = pd.melt(ebola, 
                     id_vars=['Date', 'Day'], 
                     var_name='cd_country', 
                     value_name='count')
```

```python
ebola_melt
```

```python
var_split = ebola_melt['cd_country'].str.split('_')
```

```python
status_values = var_split.str.get(0)
country_values = var_split.str.get(1)
```

```python
country_values.head()
```

```python
ebola_melt['status'] = status_values
ebola_melt['country'] = country_values
```

```python
ebola_melt.head()
```

```python
variable_split = ebola_melt['cd_country'].str.split('_', expand=True)
variable_split
```

```python
variable_split.columns = ['status1', 'country1']
variable_split.head()
```

```python
ebola_clean = pd.concat([ebola_melt, variable_split], axis=1)
```

```python
ebola_clean
```

```python
weather = pd.read_csv('./scipy-2017-tutorial-pandas-master/data/weather.csv')
weather.head()
```

```python
weather_melt = pd.melt(weather,
                      id_vars=['id', 'year', 'month', 'element'],
                      var_name='day',
                      value_name='temp')
```

```python
weather_melt.head()
```

```python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp'
).reset_index()

weather_tidy
```
